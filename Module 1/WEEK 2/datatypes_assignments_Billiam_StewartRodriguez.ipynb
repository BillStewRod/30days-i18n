```python
#
list3 = list(('hello',1,'you',6.3,'yes',7,2.3))
```

```python
# Loop through each element in the list using its index
for i in range(len(list3)):

  # If the element is a string, append 'of course'
  if str(type(list3[i])) == "<class 'str'>":
    list3[i] = list3[i] + " of course"

  # If the element is a float, add 5 to its value
  elif str(type(list3[i])) == "<class 'float'>":
    list3[i] = list3[i] + 5
```

```python
# Print the updated list after modifications
print('list3=', list3)
```
